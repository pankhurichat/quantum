```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as lin
```

```python
def kronecker(i,j):
    return 1 if i==j else 0

def V(z):
    #return A*(B*y**2-1)**2
     return a*(A*z**4-B*z**2+C)

def h(i,j):
    return (-kronecker(i+1,j) + 2*kronecker(i,j) - kronecker(i-1,j))/delta**2 + V(delta*i) * kronecker(i,j)
```

```python
A=1
B=1
C=1
a=10
b=2
delta=.01
L=6
N=300
x=np.arange(-3,3+delta,delta)
```

```python
H=np.array([[h(i,j) for i in range(-N,N+1)] for j in range(-N,N+1)])
```

```python
eigval,eigvec=lin.eig(H)
index=eigval.argsort()
eigval=eigval[index]
eigvec=eigvec[:,index]
```

```python
potential=np.array([V(delta*i) for i in range(-N,N+1)])
```

```python
plt.figure(figsize=(10,10))
for i in range(0,2):
    psi=15*eigvec[:,i]+.8*eigval[i]
    plt.plot(x,psi)
plt.plot(x,.03*potential)
plt.grid()
plt.show()
```
