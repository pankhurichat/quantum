```python
def V(x):                            #defining potential
    return (1)-(1/x)
def function(x,E):                   
    return (V(x)-E)

def y(x,E,h):
    return 1-((h**2)*(function(x,E)))/12

def numerov(x0,xf,E,n):
    psi=np.zeros(n)
    psi[0]=0
    psi[-1]=0
    psi[1]=psi[0]+h**4
    x=np.arange(x0,xf,h)
    for i in range(1,n-1):
        psi[i+1]=(psi[i]*(12-10*y(x[i],E,h))-y(x[i-1],E,h)*psi[i-1])/y(x[i+1],E,h)
    return psi
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0=.01                                       #giving boundary conditions 
xf=12
n=1000
h=(xf-x0)/n
x=np.arange(x0,xf,h)
E1=0
E2=.8
```

```python
a=numerov(x0,xf,E1,n)      #checking for bisection condition 
b=numerov(x0,xf,E2,n)
plt.plot(x,a)
plt.plot(x,b)
```

```python
def bisection(a,b,f,e):
    if f(a)*f(b)>0:
        print("enter other interval")
    while (abs(b-a)>e):
        c=(a+b)/2
        if f(a)*f(c)<0:
            b=c
        elif f(b)*f(c)<0:
            a=c
    return c
```

```python
def q(E):
    return numerov(x0,xf,E,n)[-1]
en=bisection(E1,E2,q,.01)
```

```python
plt.figure(figsize=(10,5)) 
En=numerov(x0,xf,en,n)
plt.plot(x,En)
plt.title("ground state")
plt.grid()
plt.show()
print(en)                              #printing ground energy 
```
