```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x,E):
    return 0-E

def y(x,E):
    return 1-((h**2)*f(x,E))/12
```

```python
#DEFINING THE NUMEROV ALGORITHM


def numov(x,E,n):
    psi=np.zeros(n)

    psi[0]=0
    psi[1]=psi[0]+h
    for i in range(1,n-1):
        psi[i+1]=(psi[i]*(12-10*y(x[i],E))-y(x[i-1],E)*psi[i-1])/y(x[i+1],E)
    return psi
```

```python
#INITIAL CONDITIONS

x0=-6
xf=6
n=1200
h=(xf-x0)/n
```

```python
x=np.arange(x0,xf,h)
```

```python
#INITIAL GUESSES FOR ENERGY
a=numov(x,.05,n)

e01=0.05
e02=0.09
```

```python
a
```

```python
a=numov(x,e01,n)
b=numov(x,e02,n)
```

```python
plt.plot(x,a)
plt.plot(x,b)
plt.xlabel("Position")
plt.ylabel("Energy")
plt.title("Plot for initial guesses for energy")
plt.show()
```

```python
#BISECTION METHOD TO GET ONE BOUND STATE ENERGY


def fun(e0):
    a=numov(x,e0,n)
    return a[-1]

def bi(a,b):
    c = 0                        
    while (b - a) >= 0.0001:        #tolerance 
        c = (a+b)/2                   #finding mid point

        if fun(c)*fun(a) < 0:
            b = c
        elif fun(c)*fun(b) < 0:
            a = c
        else:
            continue
    return c
```

```python
e03=bi(e01,e02)
```

```python
e03   #FIRST BOUND STATE ENERGY
```

```python
c=numov(x,e03,n)
```

```python
plt.plot(x,c)
plt.xlabel("Position")
plt.ylabel("Energy")
plt.show()
```

```python
#FUNCTION TO COUNT THE NUMBER OF NODES FOR AN ENERGY STATE

def nodes(psi):
    f=0
    for i in range(1,len(psi)-2):
        if psi[i]*psi[i+1]<0:
            f+=1
    return f
```

```python
#FUNCTION FOR FINDING ALL THE BOUND STATE ENERGIES WITHOUT MAKING GUESSES

def enval(e0,ef):
    energies=[]
    de=0.001
    n=int((ef-e0)/de)
    e1=e0+de
    for i in range(n):
        psi=numov(x,e0,1200)
        psi1=numov(x,e1,1200)
        if nodes(psi)<nodes(psi1):
            energies.append(e1)
        e0=e1
        e1=e1+de
    return energies
```

```python
s=enval(0,2)
```

```python
s
```

```python
E=[]
for i in range(len(s)):
    E.append(numov(x,s[i],n))
```

```python
for i in range(len(s)):
    plt.plot(x,E[i])
plt.xlabel("Position")
plt.ylabel("Energy")
plt.title("Plot for different bound state energies")
plt.show()
```
