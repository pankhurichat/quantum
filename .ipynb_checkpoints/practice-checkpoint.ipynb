```python
delta=0.1
L=30
N=int(L/delta)
dimension=2*N+1
x=np.linspace(-L,L,2*N+1)
```

```python
def kronecker(i,j):
    return 1 if i==j else 0

def U(y):
    return (1-np.exp(-x))**2

def h(i,j):
    return (-kronecker(i+1,j)+2*kronecker(i,j)-kronecker(i-1,j))/delta**2+U(delta*i)*kronecker(i,j)
    
```

```python
H=np.array([[h(i,j) for i in range (-N,N+1)] for j in range(-N,N+1) ])
H_eigenvalues,H_eigenvectors=lin.eig(H)
idx=H_eigenvalues.argsort()
H_eigenvalues = H_eigenvalues[idx] # Sorted eigenvalues
H_eigenvectors = H_eigenvectors[:,idx] # Sorted eigenvectors
```

```python
def psi0(y):
    return  ( 1/pow(np.pi*(a**2),0.25) ) *np.exp(-((y-b)**2)/(2.0*a**2) - 1j*p0*y)
```

```python
Psi0_m=np.sqrt(delta)*psi0(x)
plt.plot(x,Psi0_m)
```

```python
def c(i):
    coeff=np.vdot(Psi0_m,H_eigenvectors[:,i])
    return coeff

C=[]
for i in range(0,2*N+1):
    C.append(c(i))

sum=0
for i in range(len(C)):
    sum=sum+abs(C[i])**2
print(sum)
```

```python
def Psi(t): # computes the wavepacket at time tau
    summation=np.zeros((2*N+1),dtype='complex')
    for i in range(N):
        summation+= C[i]* np.exp(-1.0j*H_eigenvalues[i]*t)*H_eigenvectors[:,i]
    return summation
```

```python
def Prob(t):
    return abs(Psi(t))**2
```

# RADIAL POTENTIAL

```python
import numpy as np
from numpy import linalg as lin
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
%matplotlib inline

delta = 0.01
L = 6.
N = int(L/delta) # No. of lattice points
x = np.linspace(-6,6,2*N+1)

def kronecker(i,j):
    return 1 if i==j else 0

def V(z):
    if z!=0:
        return -2./z

def h(i,j):
    return (-kronecker(i+1,j) + 2*kronecker(i,j) - kronecker(i-1,j))/delta**2 + V(delta*i) * kronecker(i,j)


H = np.array( [[h(i,j) for i in range(1,2*N+2)] for j in range(1,2*N+2)] ) # The Hamiltonian matrix
H_eigenvalues, H_eigenvectors = lin.eig(H)
idx = H_eigenvalues.argsort() # Sorted indices (increasing order of eigenvalues)
H_eigenvalues = H_eigenvalues[idx] # Sorted eigenvalues
H_eigenvectors = H_eigenvectors[:,idx] # Sorted eigenvectors

potential = 0.01*np.array([V(delta*i) for i in range(1,2*N+2)]) # Values of U(y) stored at lattice points

psi = H_eigenvectors[:,0] # Ground state wavefunction
psi1 = H_eigenvectors[:,1]
psi2 = H_eigenvectors[:,2]

plt.plot(x, psi, label="eigenfunction- Ground State")
plt.plot(x, psi1, label="eigenfunction-First Excited State")
plt.plot(x, psi2, label="eigenfunction-Second Excited State")
plt.plot(x, 0.01*potential, label="Potential")
plt.legend();


psi.conj().T # Hermitian adjoint of psi
nc= np.dot(psi.conj().T, psi) # Normalization check
print(" Normalization check=",nc)

energy_eigenvalue = H_eigenvalues[0]
energy_eigenvalue1 = H_eigenvalues[1]
print(energy_eigenvalue)
print(energy_eigenvalue1)
```

```python
import numpy as np
from numpy import linalg as lin
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
%matplotlib inline

delta = 0.01
L = 6.
N = int(L/delta) # No. of lattice points
x = np.linspace(-6,6,2*N+1)

def kronecker(i,j):
    return 1 if i==j else 0

a=10
b=2*10**(-10)
h=6.626*10**(-34)  ## Planck constant
h_cut=h/(2*np.pi)
m=9.1*10**(-31)   ## Mass of electron (in kg)
E0=(h_cut**2)/(2*m*b**2)  ## Natural enrgy scale
alpha =  a/E0

def V(y):
    return alpha*(y**2 - 1)**2
    
def h(i,j):
    return (-kronecker(i+1,j) + 2*kronecker(i,j) - kronecker(i-1,j))/delta**2 + V(delta*i) * kronecker(i,j)


H = np.array( [[h(i,j) for i in range(-N,N+1)] for j in range(-N,N+1)] ) # The Hamiltonian matrix
H_eigenvalues, H_eigenvectors = lin.eig(H)
idx = H_eigenvalues.argsort() # Sorted indices (increasing order of eigenvalues)
H_eigenvalues = H_eigenvalues[idx] # Sorted eigenvalues
H_eigenvectors = H_eigenvectors[:,idx] # Sorted eigenvectors
```
