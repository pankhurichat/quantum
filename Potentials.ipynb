```python
a=1 #sigma
b=0 #x0
def psi0(x):
   return (1/pow(np.pi*(a**2),0.25))*np.exp(-((x-b)**2)/(2.0*a**2))
```

```python
plt.plot(x,psi0(x))
```

```python
def kronecker(i,j):
    return 1 if i==j else 0

def V(z):
    return 0

def h(i,j):
    return (-kronecker(i+1,j) + 2*kronecker(i,j) - kronecker(i-1,j))/delta**2 + V(delta*i) * kronecker(i,j)

H=np.array([[h(i,j) for i in range(-N,N+1)] for j in range(-N,N+1)])

eigval,eigvec=lin.eig(H)
index=eigval.argsort()
eigval=eigval[index]
eigvec=eigvec[:,index]
```

```python
P0=psi0(x)
def c(i):
    c= np.vdot(P0,eigvec[:,i])
    return c
```

```python
sum=0
for i in range(0,1201):
    sum+=c(i)*eigvec[:,i]
```

```python
for i in range(5):
    plt.plot(x,-eigvec[:,i])
```

```python
plt.plot(x,sum,color='red')
plt.plot(x,psi0(x),color='black')
plt.grid()
plt.show()
```

```python
def psiT(t):
    sum=0
    for i in range(0,1201):
        sum+=c(i)*np.exp(-1.0j*eigval[i]*t)*eigvec[:,i]
    return sum
```

```python
def prob(t):
    return np.array([abs(psiT(t)[i])**2 for i in range (0,1201)])
```

```python
for i in range(3):
    plt.plot(x,psiT(i))
```
